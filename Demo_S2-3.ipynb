# Matplotlib
## Importer matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np 
```

## Lineplot : plt.plot()

```python
# Création de données aléatoires sinusoïdales
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

# Création du LinePlot
plt.plot(t, s)
plt.show()
```

### Quelques paramètres utiles pour changer l'aspect

```python
# Changement de couleur
plt.plot(t, s, color="green")
plt.show()

# Changement du style de ligne
plt.plot(t, s, color="green", linestyle="--")
plt.show()
```

## Superposer deux lineplots

```python
# Création de données aléatoires sinusoïdales
t = np.arange(0.0, 2.0, 0.01)
s1 = 1 + np.sin(2 * np.pi * t)
s2 = np.cos(2 * np.pi * t)

# Création du LinePlot
plt.plot(t, s1, color='red')
plt.plot(t, s2, color='green')
plt.show()
```

```python
# Barres empilées (stacked bar plot)

# Création de données aléatoires
names = ['group_a', 'group_b', 'group_c']
values_1 = np.array([1, 10, 100])
values_2 = np.array([7, 3, 10])
values_3 = np.array([5, 8, 3])

# Création du Barplot
plt.bar(names, values_1)
plt.bar(names, values_2, bottom=values_1, color='red')
plt.bar(names, values_3, bottom=values_2+values_1, color='green')
plt.show()
```

## Histogrammes : plt.hist()

```python
# Distribution de la variable aléatoire x

# Création de données aléatoires
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# Création de l'histogramme
plt.hist(x, 50, facecolor='g')
plt.show()
```

```python
plt.hist?
```

### Transformer l'histogramme en densité de probabilité
Utilisez l'argument 'density=True' pour que la somme des valeurs dans chaque bin soit égale à 1 :

```python
# Création de l'histogramme
plt.hist(x, 50, facecolor='g', density=True)
plt.show()
```

### Ajouter une courbe de densité de probabilité

```python
# Création de données aléatoires
mu, sigma, bins = 100, 15, 50
x = mu + sigma * np.random.randn(10000)

# Création de l'histogramme
## On stocke dans trois variables différentes le résultats de plt.hist dans n, bins et patches
## On utilisera la variable bins qui correspond à notre x dans notre formule

n, bins, patches = plt.hist(x, bins, facecolor='g', density=True)

# Formule pour la densité
## Comme vu sur la figure du dessus : x suit une loi normale
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))

# Ajout d'un line plot
plt.plot(bins, y, color="r",linestyle="--")

plt.show()
```

## Les camemberts (pie charts) : plt.pie()

```python
# Pie chart où chacun des éléments va être agencé dans le sens inverse des aiguilles d'une montre
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')

plt.show()
```

```python
plt.pie?
```

### Quelques paramètres utiles

```python
# Explode

# Pie chart où chacun des éléments va être agencé dans le sens inverse des aiguilles d'une montre
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0) # Ici, on ne va faire ressortir que le Hogs

plt.pie(sizes, explode=explode,labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.show()
```

## Les boxplots : plt.boxplot()

```python
# Création de données aléatoires (en l'occurence des notes à un test de maths)
math_test = np.random.randint(1, 20+1, 30)

# Création du boxplot
plt.boxplot(math_test, labels=['Math Test'])
plt.show()
```

### Quelques paramètres utiles

```python
# Changer la limite des outliers

# Création de données aléatoires (en l'occurence des notes à un test de maths)
math_test = np.random.randint(1, 20+1, 30)

# Ajout d'outliers
math_test = np.append(math_test, np.array([22,23,24]))

# Création du boxplot
plt.boxplot(math_test,
            whis=0.6,
           labels=['Math Test'])
plt.show()
```

```python
# Ajouter la moyenne

# Création de données aléatoires (en l'occurence des notes à un test de maths)
math_test = np.random.randint(1, 20+1, 30)

# Ajout d'outliers
math_test = np.append(math_test, np.array([22,23,24]))

# Création du boxplot
plt.boxplot(math_test,
            whis=0.6,
            showmeans=True,
            meanline=True,
            labels=['Math Test'])
plt.show()
```

## Les scatter plots : plt.scatter()

```python
# Création de données aléatoires
x = np.arange(0.0, 50.0, 2.0)   # np.arange permet de créer une liste de nombres espacés d'un pas défini (ici, on va de 2 en 2, 50 fois)
y = x ** 1.3 + np.random.rand(*x.shape) * 30.0 # On crée une relation à peu près linéaire en x et y

# Création du scatter plot
plt.scatter(x, y)
plt.show()
```

### Quelques paramètres utiles

```python
# Changer la taille des marqueurs

# Création de données aléatoires
x = np.arange(0.0, 50.0, 2.0)   # np.arange permet de créer une liste de nombres espacés d'un pas défini (ici, on va de 2 en 2, 50 fois)
y = x ** 1.3 + np.random.rand(*x.shape) * 30.0 # On crée une relation à peu près linéaire en x et y


# Ajout de différentes tailles pour chacun des points
area = (15 * np.random.rand(50))**2  

# Création du scatter plot
plt.scatter(x, y, s=area)
plt.show()
```

```python
# Changer la couleur des marqueurs

# Création de données aléatoires
x = np.arange(0.0, 100, 2.0)   # np.arange permet de créer une liste de nombres espacés d'un pas défini (ici, on va de 2 en 2, 50 fois)
y = x ** 1.3 + np.random.rand(*x.shape) * 30.0 # On crée une relation à peu près linéaire en x et y


# Ajout de différentes tailles pour chacun des points
area = (15 * np.random.rand(50))**2  

# Ajout d'une séquence de couleurs
c = ["b", "r"] * 25

# Création du scatter plot
plt.scatter(x, y, s=area, c = c)
plt.show()
```

```python
# Ajouter une légende

# Création de données aléatoires
names = ['group_a', 'group_b', 'group_c']
values_1 = [1, 10, 100]
values_2 = [7, 3, 10]

# Création du Barplot
plt.bar(names, values_1)
plt.bar(names, values_2, bottom=values_1)
plt.title("Voici un titre") # Création du titre
plt.legend(["values 1", "values 2"]) # Création de légendes
plt.show()
```

```python
# Ajouter des labels aux abcisses & ordonnées

# Création de données aléatoires
names = ['group_a', 'group_b', 'group_c']
values_1 = [1, 10, 100]
values_2 = [7, 3, 10]

# Création du Barplot
plt.bar(names, values_1)
plt.bar(names, values_2, bottom=values_1)
plt.title(names) # Création du titre
plt.legend(["values_1", "values_2"]) # Création de légendes
plt.xlabel("Axe des Abscisses") # Création de labels pour les abscisses
plt.ylabel("Axe des Ordonnées") # Création de labels pour les ordonnées
plt.show()
```

## Autre façon d'utiliser matplotlib : orientée objet

```python
# Création de données
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

# Création des deux variables
fig, ax = plt.subplots()

# Construction du graphique
ax.plot(t, s) # Ajout des données en lignes

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='About as simple as it gets, folks') # Ici on fixe le titre & les lables pour les abscisses & coordonnées
plt.show()
```
